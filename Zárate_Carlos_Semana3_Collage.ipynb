```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from PIL import Image
import matplotlib.pyplot as plt
```

# **1. Se definen los filtros y kernel de paso alto**


#Filtro Sobel

```python
filtro_sobel = np.array ([[-1, 0, 1],
                           [-2, 0, 2],
                           [-1, 0, 1]])


#variable
kernel_variable = np.array([[0, 2, 0],
                             [2, -2, 2],
                             [0, 2, 0]])
```

#Filtro Laplaciano

```python
filtro_laplaciano = np.array ([[0, 1, 0],
                           [1, -4, 1],
                           [0, 1, 0]])
```

#Filtro de Marr-Hildreth (LoG):

```python
filtro_de_Marr_Hildreth = np.array ([ [0, 0,-1,-1,-1,0,0],
                                      [0,-1,-3,-3,-3,-1,0],
                                      [-1,-3,-7,-7,-7,-3,-1],
                                      [-1,-3,-7,8,-7,-3,-1],
                                      [-1,-3,-7,-7,-7,-3,-1],
                                      [0,-1,-3,-3,-3,-1,0],
                                      [0,0,-1,-1,-1,0,0]])
```

# **2. Se definen los filtros y kernel de paso bajo**

#Filtro de media (3x3):

```python
filtro_media = np.array([[1/9, 1/9, 1/9],
                         [1/9, 1/9, 1/9],
                         [1/9, 1/9, 1/9]])
```

#Filtro Gaussiano:

```python
filtro_gaussiano = np.array ([[1/16, 1/8, 1/16],
                              [1/8, 1/4, 1/8],
                              [1/16, 1/8, 1/16]])
```

# **3. Se definen los filtros y kernel direccionales**

#Filtro de Richter:

```python
filtro_de_Richter = np.array([[-3, -3, 5],
                              [-3, 0, 5],
                              [-3, -3, 5]])
```

#Filtro de Roberts en 45 grados:


```python
filtro_de_Roberts = np.array([[0, 0, 0, 0, 1],
                             [0, 0, 0,-1, 0],
                             [0, 0, 1, 0, 0],
                             [0, -1, 0, 0, 0],
                             [1, 0, 0, 0, 0]])
```

# **4. Se definen los filtros y kernel de detección de bordes**

#Filtro Operador de Prewitt (diagonal):

```python
filtro_operador_Prewitt = np.array([[1, -1, 0],
                                    [-1, 0, 1],
                                    [0, 1, 1]])
```

#Filtro Operador de Sobel (horizontal)

```python
filtro_Sobel_diagonal = np.array([[-1, -2, -1],
                                    [0, 0, 0],
                                    [1, 2, 1]])
```

# **Se hace tratamiento de la primera imagen**

```python
imagen_1 = cv2.imread('/content/drive/MyDrive/bus.jpg')


imagen_bus = cv2.resize(imagen_1, None, fx=0.3, fy=0.3, interpolation=cv2.INTER_CUBIC)


bus_gaussiano = cv2.filter2D(imagen_bus, -1, filtro_gaussiano)

bus_media = cv2.filter2D(imagen_bus, -1, filtro_media)

bus_sobel = cv2.filter2D(imagen_bus, -1, filtro_sobel)

bus_laplaciano = cv2.filter2D(imagen_bus, -1, filtro_laplaciano)

bus_Marr_Hildreth = cv2.filter2D(imagen_bus, -1, filtro_de_Marr_Hildreth)

bus_Richter = cv2.filter2D(imagen_bus, -1, filtro_de_Richter)

bus_Roberts = cv2.filter2D(imagen_bus, -1, filtro_de_Roberts)

bus_Prewitt = cv2.filter2D(imagen_bus, -1, filtro_operador_Prewitt)

bus_sobel_diagonal = cv2.filter2D(imagen_bus, -1, filtro_Sobel_diagonal)


```

#**Se crea el collage con las cuatro imágenes procesadas con filtros

Se usa Hstack y vstack para**

```python
vector = [imagen_bus, bus_gaussiano, bus_media, bus_sobel, bus_laplaciano,
          bus_Marr_Hildreth, bus_Richter, bus_Roberts, bus_Prewitt,
          imagen_cerveza, cerveza_gaussiano, cerveza_media, cerveza_sobel, cerveza_laplaciano,
          cerveza_Marr_Hildreth, cerveza_Richter, cerveza_Roberts, cerveza_Prewitt, imagen_gato,
          gato_gaussiano, gato_media, gato_sobel, gato_laplaciano,
          gato_Marr_Hildreth, gato_Richter, gato_Roberts, gato_Prewitt,
          imagen_cielo, cielo_gaussiano, cielo_media, cielo_sobel, cielo_laplaciano,
          cielo_Marr_Hildreth, cielo_Richter, cielo_Roberts, cielo_Prewitt]

filas = 4

columnas = len(vector) // filas

imagenes=[]

for nombre in vector:
  imagenes.append(nombre)

imagenes_matriz = np.array([np.array(im) for im in imagenes])

filas_imagenes = np.split(imagenes_matriz, filas)


##Se concatenan las imágenes en filas
imagenes_concatenadas_filas = [np.hstack(fila) for fila in filas_imagenes]

## Se concatenan las filas
imagenes_concatenadas_total = np.vstack(imagenes_concatenadas_filas)

plt.imshow(imagenes_concatenadas_total)
plt.axis('off')  # Desactivar los ejes
plt.show()
```
